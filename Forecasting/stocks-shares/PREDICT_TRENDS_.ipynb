```python
import pandas as pd
import numpy as np
import seaborn as sns
import time
import matplotlib.pyplot as plt
from neuralprophet import NeuralProphet
import yfinance as yf



import warnings
warnings.filterwarnings('ignore')
```

## **Read the Preprocess the Data**

```python
stock_symbol='005935.KS'
start_date='2015-01-01'
end_date='2024-01-01'
stock_data=yf.download(stock_symbol,start=start_date,end=end_date)
stock_data.head()
```

```python
stock_data.to_csv('stock_data.csv')
df=pd.read_csv('stock_data.csv')
df
```

```python
df.info()
```

```python
df['Date']=pd.to_datetime(df['Date'])
df.info()
```

```python
stocks=df[['Date','Close']]
stocks.head()
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(data=stocks,x='Date',y='Close')
plt.title('Samsung-Close',fontsize=12,weight='bold')
plt.show()
```

```python
stocks.columns=['ds','y']
stocks.head()
```

```python
stocks.info()
```

## **Train The Model**

```python
model=NeuralProphet()
model.fit(stocks)
```

## **Evaluate the Model**

```python
future=model.make_future_dataframe(stocks,periods=300)
forecast=model.predict(future)
forecast
```

```python
forecast.columns
```

```python
stocks.columns
```

```python
actual_prediction=model.predict(stocks)
plt.plot(forecast['ds'],forecast['yhat1'],label='future_prediction',c='b')
plt.plot(actual_prediction['ds'],actual_prediction['yhat1'],label='prediction Actual',c='r')
plt.plot(stocks['ds'],stocks['y'],label='actual',c='g')
plt.legend()
plt.show()
```
